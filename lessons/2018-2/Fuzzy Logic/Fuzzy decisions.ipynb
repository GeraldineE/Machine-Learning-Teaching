```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = {
    'Nombres': ['Arias', 'Benavides', 'Camargo', 'Díaz', 'Eslava'], 
    'Birth date': [1994, 1990, 1988, 1983, 1995], 
    'Starting date': [2014, 2009, 2011, 1999, 2013],
    'Salary': [1, 3, 2, 5, 4],
}
df = pd.DataFrame(data=datos)
display(df)
```

```python
def triangule(x, a, b, c):
    mf = np.zeros(shape=x.shape)
    indices = x <= a
    mf[indices] = 0
    indices = np.logical_and(x >= a, x <= b)
    mf[indices] = (x[indices]-a)/(b-a)
    indices = np.logical_and(x >= b, x <= c)
    mf[indices] = (c-x[indices])/(c-b)
    indices = x >= c
    mf[indices] = 0
    return mf

def trapezoidal(x, a, b, c, d):
    mf = np.zeros(shape=x.shape)
    indices = x <= a
    mf[indices] = 0
    indices = np.logical_and(x >= a, x <= b)
    mf[indices] = (x[indices]-a)/(b-a)
    indices = np.logical_and(x >= b, x <= c)
    mf[indices] = 1
    indices = np.logical_and(x >= c, x <= d)
    mf[indices] = (d-x[indices])/(d-c)
    indices = x >= d
    mf[indices] = 0
    return mf

variable = np.linspace(0, 6, 100)
y = trapezoidal(variable, 0, 1.5, 2.5, 3)

plt.plot(variable, y)
plt.show()
```

```python
DiscourseUniverseStartingDate = np.arange(1995,2018)
New = [2008, 2015, 2020]
Experimented = [2006, 2010, 2014]
Old = [1990, 2003, 2008]

NewMF = triangule(DiscourseUniverseStartingDate, *New)
ExperimentedMF = triangule(DiscourseUniverseStartingDate, *Experimented)
OldMF = triangule(DiscourseUniverseStartingDate, *Old)

plt.plot(DiscourseUniverseStartingDate, NewMF, label= 'NewMF')
plt.plot(DiscourseUniverseStartingDate, ExperimentedMF, label= 'ExperimentedMF')
plt.plot(DiscourseUniverseStartingDate, OldMF, label= 'OldMF')
plt.gca().legend()
plt.show()
```

```python
DiscourseUniverseSalary = np.arange(0,6)
Low = [0, 1, 2]
Medium = [1, 3, 4]
High = [3, 4, 6]

LowMF = triangule(DiscourseUniverseSalary, *Low)
MediumMF = triangule(DiscourseUniverseSalary, *Medium)
HighMF = triangule(DiscourseUniverseSalary, *High)

plt.plot(DiscourseUniverseSalary, LowMF, label= 'LowMF')
plt.plot(DiscourseUniverseSalary, MediumMF, label= 'MediumMF')
plt.plot(DiscourseUniverseSalary, HighMF, label= 'HighMF')
plt.gca().legend()
plt.show()
```

```python
DiscourseUniverseBirthDate = np.arange(1980,2018)
Young = [1997, 2003, 2018]
Medium = [1990, 1993, 2000]
Old = [1970, 1989, 1995]

YoungMF = triangule(DiscourseUniverseBirthDate, *Young)
MediumMF = triangule(DiscourseUniverseBirthDate, *Medium)
OldMF = triangule(DiscourseUniverseBirthDate, *Old)

plt.plot(DiscourseUniverseBirthDate, YoungMF, label= 'YoungMF')
plt.plot(DiscourseUniverseBirthDate, MediumMF, label= 'MediumMF')
plt.plot(DiscourseUniverseBirthDate, OldMF, label= 'OldMF')
plt.gca().legend()
plt.show()
```
